```python
import numpy as np
# fibonacci sequence
fib = [1, 2]
for num in fib:
    fib.append(fib[-1] + fib[-2])
    if len(fib) == 40:
        break 


# primonacci sequence
prim = []
for n in fib:
    prime = True
    for y in range(2, n):
        if n % y == 0:
           prime = False
           break
    if prime:
        prim.append(n)

indices = prim
print(indices)


text = list(open('password.txt').read())
text[indices[0]] = 'S'
text[indices[1]] = 'u'
text[indices[2]] = 'n'
text[indices[3]] = 'n'
text[indices[4]] = 'y'
text[indices[5]] = '='
text[indices[6]] = 'B'
text[indices[7]] = 'e'
text[indices[8]] = 's'
text[indices[9]] = 't'
with open('password.txt', 'w') as f:
    f.write("".join(text))
    password = ""
for index in indices:
    password += text[index]
print(password)
```
